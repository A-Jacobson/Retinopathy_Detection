```python
from __future__ import absolute_import
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
preds = pd.read_csv(os.path.join('..', 'submission.csv'))
```

```python
def get_labels(directory):
    return os.listdir(directory)

idx_label =  get_labels(os.path.abspath(os.path.join('E:', 'DR_Data', 'test_256', 'unknown')))
```

```python
preds['filenames'] = idx_label.values()
```

```python
del preds['Unnamed: 0']
```

```python
import numpy as np
```

```python
preds['labels'] = preds['0']
preds
```

```python
def fix_preds(x):
    if x <= 0:
        return 0.0
    elif x >= 4:
        return 4.0
    else:
        return x
```

```python
preds['labels'] = preds['0'].apply(lambda x: np.round(x, 0))
```

```python
preds['labels'] = preds['labels'].apply(fix_preds)
```

```python
preds = preds.fillna(0)
```

```python
preds['labels'] = preds['labels'].astype(int)
```

```python
preds.to_csv('answers.csv', columns=['filenames', 'labels'], index=False)
```
