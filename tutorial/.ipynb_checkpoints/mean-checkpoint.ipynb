```python
mean_r = np.mean(im[:, :, 0], dtype='uint8')
mean_g = np.mean(im[:, :, 1], dtype='uint8')
mean_b = np.mean(im[:, :, 2], dtype='uint8')
```

```python
im -= mean_r
im -= mean_g
im -= mean_b
im = im * 1/255.
```

```python
plt.imshow(im)
```

```python
def get_means(images):
    sum_r = 0 
    sum_g = 0
    sum_b = 0
    n = float(len(images))
    for im in images:
        im = misc.imread(im)
        mean_r += np.mean(im[:, :, 0])
        mean_g += np.mean(im[:, :, 1])
        mean_b += np.mean(im[:, :, 2])
    mean_r = sum_r / n
    mean_g = sum_g / n
    mean_b = sum_b / n
        
    return mean_r, mean_g, mean_b
```

```python
r, g, b = get_means(images)
```
