```python
def get_channel_means(images):
    sum_r = 0 
    sum_g = 0
    sum_b = 0
    n = float(len(images))
    for im in images:
        im = misc.imread(im)
        sum_r += np.mean(im[:, :, 0])
        sum_g += np.mean(im[:, :, 1])
        sum_b += np.mean(im[:, :, 2])
    mean_r = sum_r / n
    mean_g = sum_g / n
    mean_b = sum_b / n
    return np.array([mean_r, mean_g, mean_b], dtype='uint8')
```

```python
def get_mean(images):
    total = 0
    n = float(len(images))
    for im in images:
        im = misc.imread(im)
        total += np.mean(im)
    mean = total / n
    return np.array(mean, dtype='uint8')
    
```
